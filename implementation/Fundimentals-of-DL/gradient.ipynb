```python
"""
Experimental code
to understand how gradient values are calculated in pytorch
"""

import torch

# x = torch.tensor(1.0)
x = torch.tensor(1.0, requires_grad=True)
# y = torch.tensor(2.0)
y = torch.tensor(2.0, requires_grad=True)
# z = torch.tensor(4.0)
z = torch.tensor(4.0, requires_grad=True)
print(f'dr/dx = {x.grad}')
print(f'dr/dy = {y.grad}')
print(f'dr/dz = {z.grad}')
# forward pass
u = x*2
u.retain_grad()
v = u+y
v.retain_grad()
w = z*v
w.retain_grad()
r = w*2
r.retain_grad()
print(f'r = {r}')

# backward pass
r.backward()
print('-'*50)
print(f'dr/dr = {r.grad}')
print(f'dr/dw = {w.grad}')
print(f'dr/dv = {v.grad}')
print(f'dr/du = {u.grad}')
print()
print(f'dr/dx = {x.grad}')
print(f'dr/dy = {y.grad}')
print(f'dr/dz = {z.grad}')
print()
```
